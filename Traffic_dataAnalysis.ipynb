## 添加新label，转化对应日期为星期

```python
import calendar
from datetime import datetime

calendar.day_name[:]
```

```python
#将字符串格式转化为timestap
data_08["TRADE_DATE_DAY"]=pd.to_datetime(data_08["TRADE_DATE_DAY"],format="%Y-%m-%d")
data_09["TRADE_DATE_DAY"]=pd.to_datetime(data_09["TRADE_DATE_DAY"],format="%Y-%m-%d")
data_10["TRADE_DATE_DAY"]=pd.to_datetime(data_10["TRADE_DATE_DAY"],format="%Y-%m-%d")
data_11["TRADE_DATE_DAY"]=pd.to_datetime(data_11["TRADE_DATE_DAY"],format="%Y-%m-%d")
```

```python
#日期转星期函数
def get_weekday(dataString):
    week_day=dataString.weekday()
    return (calendar.day_name[week_day])

#获取date整列数据的星期
data_08["WEEKDAY"]=data_08["TRADE_DATE_DAY"].apply(get_weekday)
data_08.head()
```

```python
#如法炮制
data_09["WEEKDAY"]=data_09["TRADE_DATE_DAY"].apply(get_weekday) 
data_09.head()
```

```python
#如法炮制
data_10["WEEKDAY"]=data_10["TRADE_DATE_DAY"].apply(get_weekday)
data_10.head()
```

```python
#如法炮制
data_11["WEEKDAY"]=data_11["TRADE_DATE_DAY"].apply(get_weekday)
data_11.head()
```

```python
data_08["WEEKDAY"].value_counts().sort_index()
type(data_08["WEEKDAY"].value_counts())
data_08_weekday=data_08["WEEKDAY"].value_counts()
data_09_weekday=data_09["WEEKDAY"].value_counts()
data_10_weekday=data_10["WEEKDAY"].value_counts()
data_11_weekday=data_11["WEEKDAY"].value_counts()
```
