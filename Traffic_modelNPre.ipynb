```python
from datetime import timedelta

def Get_Part_of_TimeSeries(DS,TimeRange):
    return DS[TimeRange[0]:TimeRange[1]]

#均值模型（考虑每月同一天的情况）
def GenerateDs_0(TRADE_DATE_DAY):
    dayrange=pd.date_range(start=TRADE_DATE_DAY[0],end=TRADE_DATE_DAY[1],freq='1D')
    ds=pd.Series(np.zeros(len(dayrange)),index=dayrange)
    return ds

def DsList(TRADE_ADDRESS,TRADE_DATE_DAY):
    ds_list=[]
    ds=GetTimeSeries(TRADE_ADDRESS)
    for i in range(1,30):
        DayRange=TRADE_DATE_DAY-timedelta(i)
        ds_part=Get_Part_of_TimeSeries(ds,DayRange)
        #print(DayRange)
        if len(ds_part)==0 or ds_part.isnull().any():
            ds_list.append(GenerateDs_0(DayRange))
        else:
            ds_list.append(ds_part)
    return np.array(ds_list)
```

```python
#筛选掉异常的天数
#对每个月特定时间段的数据求均值和标准差，然后将均值与标准差落在10%分位数以下和90%分位数以上的日子去除
def TrueFalseListCombine(TFlist1,TFlist2):
    return [l1 and l2 for l1,l2 in zip(TFlist1,TFlist2)]

def ExceptOutlier(ds_list):
    Mean=pd.DataFrame([np.mean(i) for i in ds_list])
    mean_low=Mean > Mean.quantile(0.1)
    mean_up=Mean < Mean.quantile(0.9)
    TF=TrueFalseListCombine(mean_low.values,mean_up.values)
    mean_index=Mean[TF].index.values
    Std=pd.DataFrame([np.std(i) for i in ds_list])
    std_low=Std > Std.quantile(0.1)
    std_up=Std < Std.quantile(0.9)
    TF=TrueFalseListCombine(std_low.values,std_up.values)
    std_index=Std[TF].index.values
    valid_index=list(set(mean_index)&set(std_index))
    return valid_index
    #return ds_list
```

```python
TRADE_ADDRESS_list=[121,123,125,127,129,131,133,135,137,139,141,143,145,147,149,151,153,155,157,159]
```

```python
def Predict(TrainTime,PredictTime):
    count=0
    tag=TRADE_ADDRESS_list[0]
    
    Predict=Do_ARMA(tag,TrainTime,PredictTime,2,0,Draw=False)
    
    def TrainsTime(time):
        date=str(time.date())
        output=date
        return output
    
    date_p=[TrainsTime(time) for time in Predict.index]
    passengerCount=Predict.values
    TRADE_ADDRESS=[tag]*len(Predict)
    Predict_Result=pd.DataFrame({'Trade Address':TRADE_ADDRESS,'日期':date_p,'预测客流量':passengerCount})
    Predict_Result=Predict_Result[['日期','Trade Address','预测客流量']]
    
    for tag in TRADE_ADDRESS_list[1:]:
        Predict=Do_ARMA(tag,TrainTime,PredictTime,2,0,Draw=False)
        date_p=[TrainsTime(time) for time in Predict.index]
        passengerCount=Predict.values
        TRADE_ADDRESS=[tag]*len(Predict)
        Predict_Result_Part=pd.DataFrame({'Trade Address':TRADE_ADDRESS,'日期':date_p,'预测客流量':passengerCount})
        Predict_Result_Part=Predict_Result_Part[['日期','Trade Address','预测客流量']]
        Predict_Result=pd.concat([Predict_Result,Predict_Result_Part])
        count+=1
        #print(count)
        
    Predict_Result['预测客流量']=np.nan_to_num(Predict_Result['预测客流量'])
    return Predict_Result
```
